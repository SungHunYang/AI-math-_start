The Graphs of Odds and Logit

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn')

p_np = np.linspace(0.01,0.99, 100)
p_tf = tf.linspace(0.01,0.99,100)

odds_np = p_np/(1-p_np)
odds_tf = p_tf/(1-p_tf)

logit_np =np.log(odds_np)
logit_tf = tf.math.log(odds_tf)

fig, axes = plt.subplots(2,1,figsize=(15,10),
                         sharex= True)

axes[0].plot(p_np,odds_np)
axes[1].plot(p_np,logit_np)

xticks = np.arange(0,1.1,0.2)
axes[0].tick_params(labelsize=15) 
axes[0].set_xticks(xticks)
axes[0].set_ylabel('Odds',fontsize=20,color='darkblue')
axes[1].tick_params(labelsize=15) 
axes[1].set_xticks(xticks)
axes[1].set_ylabel('logit',fontsize=20,color='darkblue')
axes[1].set_xlabel('probablity',fontsize=20,color='darkblue')
```

The Graphs of Sigmoid

```python
import tensorflow as tf
from tensorflow.keras.layers import Activation

x=tf.linspace(-30,30,100)
sigmoid = Activation('sigmoid')(x)

fig, ax = plt.subplots(figsize =(10,5))
ax.plot(x.numpy(),sigmoid.numpy())

# sigmoid는 잘 보면 0 아니면 1만 출력하게 만든다.
```

Single-variate Logistic Regression Models

```python
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.layers import Dense

plt.style.use('seaborn')

X = tf.random.normal(shape=(100,1))

dense = Dense(units=1, activation = 'sigmoid')

Y = dense(X)

fig, ax = plt.subplots(figsize=(7,7))
ax.scatter(X.numpy().flatten(), Y.numpy().flatten())
# 학습이 끝나면 1 또는 0에 거의 값이 모일 것이다.
```

Multi-variate Logistic Regression Models

```python
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.layers import Dense

plt.style.use('seaborn')

X = tf.random.normal(shape=(100,5)) # 여러 종류가 추가되어 multi-variate 가 된다.

dense = Dense(units=1, activation = 'sigmoid') # 뉴런이 1개 이기 때문에 결국 binary가 결과로 나온다.

Y = dense(X)
print(Y.shape)
```

Binary Classifier with Dense Layers

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential

'''
이 모델이 바로 binary Classifier가 된다

모델이 보면 Sequential 하게 3가지 layer가 들어왔는데
마지막 layer가 뉴런이 1 개이고 sigmoid 형태이기 때문에 결과가
확률의 형태로 1개 나오는 binary Classifier가 된다.

'''

model = Sequential()

model.add(Dense(units=10, activation='relu'))
model.add(Dense(units=5, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))
```
